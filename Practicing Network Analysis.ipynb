```python
# http://konect.uni-koblenz.de/networks/cit-HepTh
G2 = nx.read_adjlist("cit-HepTh/out.cit-HepTh") 
```

```python
nodes_G2 = nx.nodes(G2)
edges_G2 = nx.edges(G2)
directed = nx.is_directed(G2)
print(len(nodes_G2),len(edges_G2), directed)
```

```python
nbunch = nodes_G2[0:500]
H = G2.subgraph(nbunch)
```

```python
nx.draw(H)
plt.show()
```

```python
cH = nv.CircosPlot(H)
cH.draw()
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.star_graph(20)
pos = nx.spring_layout(G)
colors = range(20)
nx.draw(G, pos, node_color='#A0CBE2', edge_color=colors,
        width=4, edge_cmap=plt.cm.Blues, with_labels=False)
plt.show()
```

```python
import networkx as nx
G=nx.Graph()
G.add_edges_from([(1,2),(2,3),(3,1),(1,4)]) #define G
fixed_positions = {1:(0,0),2:(-1,2)}#dict with two of the positions set
fixed_nodes = fixed_positions.keys()
pos = nx.spring_layout(G,pos=fixed_positions, fixed = fixed_nodes)
nx.draw_networkx(G,pos)
plt.show()
```
